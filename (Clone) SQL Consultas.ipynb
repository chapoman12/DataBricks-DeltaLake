```python
%sql
USE CATALOG hive_metastore
```

```python
%sql
CREATE TABLE employees (
  id INT, name STRING, salary DOUBLE
)
```

```python
%sql
INSERT INTO employees VALUES (1, "Adam", 3500.0),
(2, "Sarah", 4020.5);

INSERT INTO employees VALUES (3, "John", 2999.3),
(4, "Tomas", 35000.3);

INSERT INTO employees VALUES (5, "Sally", 3500.0),
(6, "Mary", 4020.5);
```

```python
%sql
SELECT * FROM employees
```
